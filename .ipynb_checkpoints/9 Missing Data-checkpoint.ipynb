```python
import pandas as pd
df=pd.read_json('data.json')
df
```

```python
df.dropna()   # all rows having NaN values dropped, axis=0 is by default
```

```python
df.dropna(axis=1)     # all columns having NaN values dropped
```

```python
df.dropna(how='any')  # this command is used to drop all rows having all NaN values in comlete row
```

```python
df.dropna(how='all')
```

```python
df.dropna(subset=['Duration'])   # drop NaN values in Duration column but NaN values of other columns will not be dropped
```

```python
df.dropna(inplace=True)
df
```

```python
df.dropna(thresh=1)
```

```python
df.fillna('python')  # all NaN values replaced(filled) with python
```

```python
df.fillna({'Duration': 'python', 'Pulse': 'java'})  # all NaN values will be filled by python in duraion column and with java in pulse column
```

```python
df.fillna(method='ffill') # 2 methods ffill means forward fill and bfill means.....
```

```python
df.fillna(method='ffill', axis=1)  # column wise
```

```python
df.fillna(12, inplace =True)      # NaN is replaced by 12
```

```python
df.fillna('python', limit=2) # NaN replaced with python but only for 2 NaN values 3rd will nort replaced in a column 
```
