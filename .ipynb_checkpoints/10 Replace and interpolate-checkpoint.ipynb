# Replace

```python
import pandas as pd
df=pd.read_json('data.json')
df
```

```python
df.replace(to_replace=60, value=45) # jis jis jga p value 60 ha waha 45 ho jaye gi
```

```python
df.replace([60,60,60,45,45],20) # pehlay 5  ki jga 20 a jye ga
```

```python
df.replace('A-Za-z', 'python', regex=True) # all alphabet capital and small replaced with python 
```

```python
df.replace({'Duration':'[A-Z]'},22, regex=True) # it will replace all capital alphabets in column duraion with 22 (if any)
```

```python
df.replace(1,method='ffill')  # jis column ma jis jga 1 ha forward walay se replace kr dia 1 ko and same for bfill
```

```python
df.replace(1,method='ffill', limit=2)  # sirf pehlay 2 replaced 
```

```python
df.replace(1,method='ffill', limit=2, inplace=True)
```

```python
# inplace=True means all changings in orignal data not copy, next time changed data will be displayed always
```

# Interpolate


```python
df
```

```python
df.interpolate() # All NaN values filled automatically (linearly) only for numeric values not for string
```

```python
df.interpolate(method='linear') # many other methods but linear method is by default
```

```python
df.interpolate(method='linear', axis=0) # column wise
```

```python
df.interpolate(method='linear', axis=1) # row wise lakin data type sb ka same ha phr ye 
# work kray ga ager kuch strings hn aur kuch numeric error a jaye ga
```

```python
df.interpolate(method='linear', axis=0, limit=2) # sirf pehlay 2 NaN p apply ho ga
```

```python
df.interpolate(method='linear', axis=0, limit=2, limit_direction='forward') # pehli 2 backward means last 2 , 
# both means pehli and last 2,2
```

```python
df.interpolate(limit_area='inside') # interpolate all NaN values and 'outside' ma nahi kray ga NaN values as it is hun gi
```
