# Join

```python
import pandas as pd
df1=pd.DataFrame({'A':[1,2,3,4], 'B':[5,6,7,8]})
df2=pd.DataFrame({'C':[10,20,30,40], 'D':[55,66,77,88]})
df1.join(df2)
```
