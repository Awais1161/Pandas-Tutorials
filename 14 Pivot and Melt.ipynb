# Melt

```python
import pandas as pd
df=pd.DataFrame({'days':[1,2,3,4,5,6], 'eng': [10,12,14,12,15,14], 'maths': [11,14,15,13,12,16]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=['eng']) # eng used just like index
```

```python
pd.melt(df,id_vars=['days'])
```

```python
pd.melt(df,id_vars=['days'], var_name='python')
```

```python
pd.melt(df,id_vars=['days'], var_name='python', value_name='abc')
```

# Pivot

```python
df=pd.DataFrame({'days':[1,2,3,4,5,6], 'st_name':['a', 'b', 'c', 'a', 'b', 'c'],
                 'eng': [10,12,14,12,15,14], 'maths': [11,14,15,13,12,16]})
df
```

```python
df.pivot(index='days', columns='st_name') # df.pivot() will give error as atleast 2 parametrs are required
```

```python
df=pd.DataFrame({'days':[1,2,3,4,5,6], 'st_name':['a', 'b', 'a', 'a', 'b', 'b'],
                 'eng': [10,12,14,12,15,14], 'maths': [11,14,15,13,12,16]})

```

```python
df.pivot(index='days', columns='st_name')
```

```python
df.pivot(index='days', columns='st_name', values='eng') # only english
```

```python
df=pd.DataFrame({'days':[1,1,1, 1,2,2], 'st_name':['a', 'b', 'a', 'a', 'b', 'b'],
                 'eng': [10,12,14,12,15,14], 'maths': [11,14,15,13,12,16]})
df
```

```python
df.pivot_table(index='st_name', columns='days',aggfunc='mean')
```

```python
df.pivot_table(index='st_name', columns='days')
```

```python
df.pivot_table(index='st_name', columns='days',aggfunc='sum', margins='False')
```

```python
df.pivot_table(index='st_name', columns='days',aggfunc='sum', margins='True')
```

```python
df.pivot_table(index='st_name', columns='days',aggfunc='mean', margins='True')
```

```python
df.pivot_table(index='st_name', columns='days',aggfunc='mean')
```
